## Initial Data Inspection

### Subtask:
Inspect the data types of each column and re-confirm the absence of missing values to ensure data quality and readiness for analysis. Additionally, check for unique values in categorical columns.


**Reasoning**:
To inspect the data types of each column, I will use the `.info()` method, which provides a concise summary of the DataFrame including column names, non-null values, and data types.



```python
print(df.info())
```

**Reasoning**:
To reconfirm the absence of missing values, I will use `df.isnull().sum()` which will show the count of null values for each column.



```python
print('Missing values per column:\n', df.isnull().sum())
```

**Reasoning**:
To understand the categories within each categorical column, I will print the unique values for 'Sex', 'Chest pain type', 'FBS over 120', 'EKG results', 'Exercise angina', 'Slope of ST', 'Number of vessels fluro', 'Thallium', and 'Heart Disease'.



```python
categorical_cols = ['Sex', 'Chest pain type', 'FBS over 120', 'EKG results', 'Exercise angina', 'Slope of ST', 'Number of vessels fluro', 'Thallium', 'Heart Disease']

for col in categorical_cols:
    print(f"Unique values for '{col}': {df[col].unique()}")
```

## Descriptive Statistics

### Subtask:
Calculate and display descriptive statistics (mean, median, standard deviation, min, max, etc.) for numerical columns to understand their central tendency, dispersion, and shape of distribution.


**Reasoning**:
To calculate and display descriptive statistics for the specified numerical columns, I will use the `.describe()` method on the DataFrame, selecting only the relevant columns.



```python
numerical_cols = ['Age', 'BP', 'Cholesterol', 'Max HR', 'ST depression']
descriptive_stats = df[numerical_cols].describe()
print('Descriptive statistics for numerical columns:\n', descriptive_stats)
```

## Distribution of Numerical Features

### Subtask:
Visualize the distribution of key numerical features ('Age', 'BP', 'Cholesterol', 'Max HR', 'ST depression') using histograms and box plots to identify their patterns, outliers, and skewness. Ensure plots have appropriate labels and titles.


**Reasoning**:
To visualize the distribution of each numerical feature, I will iterate through the `numerical_cols` list and generate a histogram and a box plot for each feature, ensuring clear titles and labels for better interpretability.



```python
for col in numerical_cols:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    sns.histplot(df[col], kde=True, ax=axes[0])
    axes[0].set_title(f'Distribution of {col}')
    axes[0].set_xlabel(col)

    sns.boxplot(y=df[col], ax=axes[1])
    axes[1].set_title(f'Box Plot of {col}')
    axes[1].set_ylabel(col)

    plt.tight_layout()
    plt.show()
```
