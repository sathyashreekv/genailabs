```python
from transformers import pipeline

# Load a pre-trained sentiment analysis model
sentiment_pipeline = pipeline("sentiment-analysis")

# Analyze the sentiment of some text
text = "I love using Hugging Face! It makes NLP so much easier."
result = sentiment_pipeline(text)

print(f"Text: {text}")
print(f"Sentiment: {result[0]['label']}, Score: {result[0]['score']:.2f}")

text2 = "This is a terrible movie, I regret watching it."
result2 = sentiment_pipeline(text2)

print(f"\nText: {text2}")
print(f"Sentiment: {result2[0]['label']}, Score: {result2[0]['score']:.2f}")
```

## Define Prediction Function

### Subtask:
Create a Python function that takes text as input, uses the loaded sentiment analysis pipeline to predict the sentiment, and returns the result (label and score).


**Reasoning**:
To fulfill the subtask of defining a sentiment prediction function, I will add a code cell containing the Python function `sentiment_predictor` as described in the instructions. This function will utilize the previously loaded `sentiment_pipeline`.



```python
def sentiment_predictor(text):
    """
    Predicts the sentiment (label and score) of a given text using the pre-loaded sentiment_pipeline.
    """
    result = sentiment_pipeline(text)
    label = result[0]['label']
    score = result[0]['score']
    return label, score

print("Sentiment prediction function 'sentiment_predictor' defined.")
```

## Create Gradio Interface

### Subtask:
Design the Gradio web interface by linking the prediction function to Gradio's `Interface` class, specifying input components (text box) and output components (text for label, number for score).


**Reasoning**:
To design the Gradio interface, I will create a `gr.Interface` instance, linking it to the `sentiment_predictor` function and defining the input as a `gr.Textbox` and outputs as a `gr.Textbox` for the label and a `gr.Number` for the score, along with a title.



```python
sentiment_interface = gr.Interface(
    fn=sentiment_predictor,
    inputs=gr.Textbox(lines=5, label='Text Input'),
    outputs=[gr.Textbox(label='Sentiment Label'), gr.Number(label='Sentiment Score')],
    title='Sentiment Analysis with Hugging Face and Gradio'
)

print("Gradio interface 'sentiment_interface' defined.")
```
